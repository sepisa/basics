```python
import numpy as np
import matplotlib.pyplot as plt
```

## uniform distribution

```python
values = np.random.uniform(-10.0, 10.0, 10000)
plt.hist(values, 50)
plt.show()
```

## Normal / Gussian

A normal continuous random variable.

 norm.pdf(x, loc, scale)

The location (loc) keyword specifies the mean. The scale (scale) keyword specifies the standard deviation.

```python
from scipy.stats import norm

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

generate some random number with normal distribution."mu" is the desired mean and "sigma" is the variance

```python
sigma=2.0
mu=5.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

exponential pdf / power low 

```python
from scipy.stats import expon

values = np.arange(0, 10, 0.001)
plt.plot(values, expon.pdf(values))
```

# binomial probability Mass funcion

```python
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

```python
from scipy.stats import poisson
mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
